```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)
roberts_x = np.array([[1, 0], [0, -1]], dtype=np.float32)
roberts_y = np.array([[0, 1], [-1, 0]], dtype=np.float32)

gradient_x = cv2.filter2D(image, -1, roberts_x)
gradient_y = cv2.filter2D(image, -1, roberts_y)

plt.figure(figsize=(20,8))
plt.subplot(141),plt.imshow(image,cmap='gray'),plt.title('Original'),plt.axis('off')
plt.subplot(142),plt.imshow(gradient_x ,cmap='gray'),plt.title('gradient_x'),plt.axis('off')
plt.subplot(143),plt.imshow(gradient_y,cmap='gray'),plt.title('gradient_y'),plt.axis('off')
plt.subplot(144),plt.imshow(gradient_x + gradient_y,cmap='gray'),plt.title('Roberts'),plt.axis('off')
plt.show()
```

```python
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
gradient_x = cv2.filter2D(image, -1, kernelx)
gradient_y = cv2.filter2D(image, -1, kernely)

plt.figure(figsize=(20,8))
plt.subplot(141),plt.imshow(image,cmap='gray'),plt.title('Original'),plt.axis('off')
plt.subplot(142),plt.imshow(gradient_x ,cmap='gray'),plt.title('gradient_x'),plt.axis('off')
plt.subplot(143),plt.imshow(gradient_y,cmap='gray'),plt.title('gradient_y'),plt.axis('off')
plt.subplot(144),plt.imshow(gradient_x + gradient_y,cmap='gray'),plt.title('perwitt'),plt.axis('off')
plt.show()
```

```python
gradient_x= cv2.Sobel(image,cv2.CV_64F,1,0,ksize=5)
gradient_x = cv2.convertScaleAbs(gradient_x)
gradient_y = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=5)
gradient_y = cv2.convertScaleAbs(gradient_y)
plt.figure(figsize=(20,8))
plt.subplot(141),plt.imshow(image,cmap='gray'),plt.title('Original'),plt.axis('off')
plt.subplot(142),plt.imshow(gradient_x ,cmap='gray'),plt.title('gradient_x'),plt.axis('off')
plt.subplot(143),plt.imshow(gradient_y,cmap='gray'),plt.title('gradient_y'),plt.axis('off')
plt.subplot(144),plt.imshow(gradient_x + gradient_y,cmap='gray'),plt.title('sobel'),plt.axis('off')
plt.show()
```

```python
import cv2
image = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(image, 100,200)

plt.figure(figsize=(20,8))
plt.subplot(141),plt.imshow(image,cmap='gray'),plt.title('Original'),plt.axis('off')
plt.subplot(142),plt.imshow(edges,cmap='gray'),plt.title('canny'),plt.axis('off')
```

```python
image = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)
img_gaussian = cv2.GaussianBlur(image,(5,5),0)
edges =cv2.Laplacian(img_gaussian, cv2.CV_64F)
dest = cv2.convertScaleAbs(edges)

plt.figure(figsize=(20,8))
plt.subplot(141),plt.imshow(image,cmap='gray'),plt.title('Original'),plt.axis('off')
plt.subplot(142),plt.imshow(dest,cmap='gray'),plt.title('laplacian'),plt.axis('off')
```
